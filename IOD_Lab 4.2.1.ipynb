```python
# Predictor matrix - 2D array
X = wine[['alcohol', 'free sulfur dioxide', 'pH']]
X
```

### 4. Using Linear Regression Create a Model and Test Score

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# Train-Test Split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
# Create a model for Linear Regression
# Fit the model with the Training data
model = LinearRegression().fit(X_train, y_train)
print(model.intercept_)
print(model.coef_)

# Calculate the score (R^2 for Regression) for Training Data
print(model.score(X_train, y_train))

# Calculate the score (R^2 for Regression) for Testing Data
# Although you can use x_train and y_train to check the goodness of fit, this isnâ€™t a best practice
#An unbiased estimation of the predictive performance of your model is based on test data
print(model.score(X_test, y_test))
```

```python
# Set up 5-fold cross validation 
#https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.KFold.html#sklearn.model_selection.KFold
#https://www.youtube.com/watch?v=TIgfjmp-4BA&ab_channel=Udacity
#is a way to use "all" data for training and testing to enhance accuracy since there is tradeoff between train and test split

k_fold = KFold(n_splits=5, shuffle=True) #n_splits = no. of folds
train_scores = []
train_rmse = []
test_scores = []
test_rmse = []

regr = LinearRegression() #instantiate model object
#do this because everytime you write "LinearRegression()" it creates a new object and forgets what you previously fit

for k, (train, test) in enumerate(k_fold.split(X)):
    # use enumerate() anytime you need to use the count and an item in a loop
    # k_fold.split(data) returns generator object, so u use it in a for...in...statement
    
    # Get training and test sets for X and y
    Xtrain = X.loc[train]
    ytrain = y.loc[train]
    Xtest = X.loc[test]
    ytest = y.loc[test]
    
    # Fit model with training set
    model = regr.fit(Xtrain, ytrain)
    
    # Make predictions with training and test set
    ypred_train = regr.predict(Xtrain)
    ypred_test = regr.predict(Xtest)
    
    # Score R2 and RMSE on training and test sets and store in list
    train_scores.append(regr.score(Xtrain, ytrain))
    test_scores.append(regr.score(Xtest, ytest))
    train_rmse.append(mean_squared_error(ytrain, ypred_train))
    test_rmse.append(mean_squared_error(ytest, ypred_test))
    
# Create a metrics_df dataframe to display r2 and rmse scores
metrics_df = pd.DataFrame(list(zip(train_scores, test_scores, train_rmse, test_rmse)),
               columns =['train scores', 'test scores', 'train RMSE', 'test RMSE'])
metrics_df
```
