# Statistics - Part 1

```python
import numpy as np
import matplotlib.pyplot as plt
# get matplot to show charts inline
%matplotlib inline

# %matplotlib inline is: magic function that forces matplotlib to keep charts within same notebook instead of opening charts in separate windows
```

```python
# plot frequency of some observations using pie chart
labels = ['Developers', 'Data Engineers', 'Data Scientists', 'Business Analysts']
frequency = [15, 32, 12, 7]
fig1, ax1 = plt.subplots()
# subplots takes 3 arguments: rows, columns, index of current plot (to produce figure)

explode = (0.3, 0, 0, 0.4)  # only "explode" the 3rd slice (i.e. 'Data Scientist')
ax1.pie(frequency, explode=explode, labels=labels, radius=1, autopct='%1.2f%%',
        shadow=True, startangle=90)
# autopct formats displayed percentage as 1d.p.
# if you do 1.2, it displays 2d.p.

# ax1.axis('equal')
# Equal aspect ratio ensures that pie is drawn as a circle.
# makes radius = 1 by default, if you want to adjust radius, need to remove this line

plt.show()
```

```python
# SAME CODE BUT use plt.figure(figsize()) to adjust piechart size
labels = ['Developers', 'Data Engineers', 'Data Scientists', 'Business Analysts']
frequency = [15, 32, 12, 7]

plt.figure(figsize=(10,10),dpi=50)
explode = (0.3, 0, 0, 0.4)  # only "explode" the 3rd slice (i.e. 'Data Scientist')
plt.pie(frequency, explode=explode, labels=labels, radius=1, autopct='%1.1f%%',
        shadow=True, startangle=90, colors=('#AB466E', '#780D38', '#C872DF', '#E1E2AC'))
plt.show()
```

```python
# plot frequency of some observations using bar chart
objects = ('Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp')
x_pos = np.arange(len(objects))
usage = [10,8,6,4,2,3]
 
# Create a bar plot
# plt.bar(x, height, width, bottom, align)
plt.bar(objects, usage, width=1, align='edge',
        color=('#e63e73','#eb9634','#ebd334','#59c73e','#2eabd9','#db56f0'))

# plt.xticks takes arguments: ticks, labels
plt.xticks(x_pos, objects)

plt.ylabel('Usage')
plt.title('Programming language usage')
 
plt.show()
```
