```python
# Practice the use of matplotlib
import numpy as np
import matplotlib.pyplot as plt
# Create a data set 
# 100 evenly spaced values
x = np.linspace(0, 10, 100)
# Print the data
print(x)
# Create a figure using plt.figure
# if you wish to plot more than one graph on same plot, need to include this line
plt.figure()
# Plot the sine wave of x
plt.plot(x, np.sin(x))
# Plot the cosine wave of x
plt.plot(x, np.cos(x))
plt.show()
```

```python
# Create two arrays of 5 arbitrary numbers 
array_1 = [5,7,8,3,6]
array_2 = [3,8,4,2,2]
# Plot two bar charts using plt.bar and the arrays you've created above
# Label them "Example one" and "example two"
plt.bar(np.arange(len(array_1)), height=array_1, width=0.5, label="Example one")
plt.bar(np.arange(len(array_2)) +0.25, height=array_2, width=0.5, label="Example two", color='g')
plt.legend()
# Label x axis "Bar number"
plt.xlabel('Bar number')
# Label y axis "Bar height"
plt.ylabel('Bar height')
# Set title "Bar Graph"
plt.title("Bar Graph")
# Show the chart
plt.show()
```

```python
# Histogram Code 
# Create an array of a random normal distribution
sample_dist = np.random.laplace(loc=15, scale=3, size=100)
# Create a histrogram using plt hist()
plt.style.use('grayscale')
n, bins, patches = plt.hist(x=sample_dist, bins='auto',
                            alpha=0.7, rwidth=0.75)
# Set grid for x and y
# set transperancy of the grid using alpha=0.5
plt.grid(True, alpha=0.5)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('My First Histogram Ever')
plt.text(5, 15, r'$\mu=15, b=3$')


maxfreq = n.max()
# Set a clean upper y-axis limit.p
plt.ylim(ymax=np.ceil(maxfreq / 10) * 10 if maxfreq % 10 else maxfreq + 10)
# "if maxfreq % 10" means "if maxfreq is perfectly divisible by 10 with no remainder"
# if maxfreq is multiple of 10, add 10 to y upper bound so there's a nice space lol
# in this case since maxfreq is 23, it will just round up to 30. if maxfreq instead was 30, it will round up to 40.
```
